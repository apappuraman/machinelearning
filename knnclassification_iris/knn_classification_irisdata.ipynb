```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
data=load_iris()
```

```python
data
```

```python
x=data.data
```

```python
y=data.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
model=KNeighborsClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
df_comp=pd.DataFrame({"actual": y_test, "predicted": y_pred})
```

```python
df_comp.head()
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("mse", mse)
print("mae", mae)
print("r2", r2)
```

```python
model2=KNeighborsClassifier(n_neighbors=15)
model2.fit(x_train, y_train)
```

```python
y2_pred=model2.predict(x_test)
```

```python
mse2 = mean_squared_error(y_test, y2_pred)
mae2 = mean_absolute_error(y_test, y2_pred)
r22 = r2_score(y_test, y2_pred)

print("mse", mse2)
print("mae", mae2)
print("r2", r22)
```

```python
result1 = confusion_matrix(y_test, y2_pred)
print("Confusion Matrix:")
print(result1)
result2 = classification_report(y_test, y2_pred)
print("Classification Report:",)
print (result2)
result3 = accuracy_score(y_test,y2_pred)
print("Accuracy:",result3)
```
